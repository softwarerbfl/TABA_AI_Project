# 결과 시각화

Batch size, Learning rate에 따른 test accuracy, recall, precision, F1-score 값 시각화 

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
learning_rate=[0.0001, 0.001]
batch_size=[16,24,32,40]
```

# Optimizer: Adam 사용

```python
test_acc_1=[0.7305, 0.7365, 0.7335, 0.738]
test_recall_1=[ 0.778, 0.751, 0.75, 0.684]
test_precision_1=[ 0.711, 0.7298, 0.726, 0.769]
test_f1_1=[0.743, 0.7403, 0.738, 0.723]
```

```python
test_acc_2=[0.7105, 0.6955, 0.7175, 0.7205]
test_recall_2=[0.616, 0.649, 0.685, 0.688]
test_precision_2=[0.7596, 0.716, 0.7326, 0.7358]
test_f1_2=[0.68, 0.681, 0.708, 0.711]
```

```python
values=range(len(batch_size))
#학습률 0.0001
fig=plt.figure(figsize=(12,8))
plt.title('learning rate=0.0001')
plt.xlabel('Batch Size')
plt.ylabel('Value')
plt.plot(values, test_acc_1,marker='o',markersize=6,label='accuracy')
plt.plot(values, test_recall_1,marker='o',markersize=6, label='recall')
plt.plot(values, test_precision_1,marker='o',markersize=6, label='precision')
plt.plot(values, test_f1_1,marker='o',markersize=6, label='F1-score')
plt.ylim([0.6, 0.9])
plt.xticks(values,batch_size )
plt.legend(fontsize = 16)
plt.show()
```

```python
#학습률 0.001
fig=plt.figure(figsize=(12,8))
plt.title('learning rate=0.001')
plt.xlabel('Batch Size')
plt.ylabel('Value')
plt.plot(values, test_acc_2,marker='o',markersize=6,label='accuracy')
plt.plot(values, test_recall_2,marker='o',markersize=6, label='recall')
plt.plot(values, test_precision_2,marker='o',markersize=6, label='precision')
plt.plot(values, test_f1_2,marker='o',markersize=6, label='F1-score')
plt.ylim([0.6, 0.9])
plt.xticks(values,batch_size )
plt.legend(fontsize = 16)
plt.show()
```

# 최종 모델 성능 비교(Accuracy)
